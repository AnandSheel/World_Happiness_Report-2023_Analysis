```python
data_columns = ['Country name', 'Regional indicator', 'Happiness score', 'Logged GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']
```

```python
data = data[data_columns].copy()
```

```python
# Renaming the columns to make it more simpler and easy to understand
```

```python
mydict = {'Country name':'country_name', 'Regional indicator':'regional_indicator', 'Happiness score':'happiness_score', 'Logged GDP per capita':'logged_GDP_per_capita', 'Social support':'social_support', 'Healthy life expectancy':'healthy_life_expectancy', 'Freedom to make life choices':'freedom_to_make_life_choices', 'Generosity':'generosity', 'Perceptions of corruption':'perceptions_of_corruption'}
```

```python
happy_df = data.rename(columns = mydict)
```

```python
# Check if any column have null values
happy_df.isnull().sum()
```

```python
happy_df.shape
```

```python
# drop the null value
happy_df.dropna(inplace = True)
```

```python
# recheck for null values
happy_df.shape
```

### Comparison between Happiness score  and GDP for different regions 

```python
# plot b/w happiness and GDP

plt.rcParams['figure.figsize'] = (15, 7)
plt.title('Plot between Happiness Score and GDP')
sns.scatterplot( x = happy_df.happiness_score, y = happy_df.logged_GDP_per_capita, hue = happy_df.regional_indicator, s = 200)

plt.legend( loc = 'upper left', fontsize = '10')
plt.xlabel('Happiness Score')
plt.ylabel('GDP per capita')
```

### GDP by Region

```python
gdp_region = happy_df.groupby('regional_indicator')['logged_GDP_per_capita'].sum()
gdp_region
```

```python
# pieplot
gdp_region.plot.pie(autopct = '%1.1f%%')  # autopct = '%1.1f%%' - formats the percentage to one decimal place
plt.title('GDP by Region')
plt.ylabel(' ')
```

```python
count_of_countries = happy_df.groupby('regional_indicator')[['country_name']].count()
print(count_of_countries)
```

### correlation map

```python
corr = happy_df.corr(numeric_only=True)
fig = plt.figure(figsize = (12,8))
sns.heatmap(corr, cmap = 'Blues', linewidth = 0, annot = True)
```

-This heat map reveals several strong correlations between factors and happiness score. It has been found that countries with higher GDP per capita, healthier life expectancies, greater freedom to make choices, and stronger social support tend to have higher happiness scores. Furthermore, a strong positive correlation has been found between GDP per capita, healthy life expectancy, and social support.

-In addition, a negative correlation has been found between happiness score and perceptions of corruption. This suggests that maintaining high levels of happiness among citizens may be challenging for countries with higher levels of corruption.

-Overall, these findings highlight the importance of economic, social, and political factors in determining happiness levels across countries.

### Corruption in different Regions

```python
# corruption in regions

corruption = happy_df.groupby('regional_indicator')[['perceptions_of_corruption']].mean()
corruption
```

```python
# bar plot

plt.rcParams['figure.figsize'] = (12,8)
plt.title('Perceptions of Corruption in various Regions')
plt.xlabel('Regions', fontsize = 15)
plt.ylabel('Corruption Index', fontsize = 15)
plt.xticks(rotation = 30, ha = 'right')         # for the rotation angle of x-axis
plt.bar(corruption.index, corruption.perceptions_of_corruption)
```

## Top 10 and Bottom 10 Countries

```python
top_10 = happy_df.head(10)

bottom_10 = happy_df.tail(10)
```

### In terms of happiness

```python
fig, ax = plt.subplots(2, 1, figsize = (15,10))

sns.barplot(top_10, y = 'happiness_score', x = 'country_name', palette = 'tab20', ax = ax[0])
ax[0].set_title("Top 10 happiest countries")



sns.barplot(bottom_10, y = 'happiness_score', x = 'country_name', palette = 'tab20', ax = ax[1])
ax[1].set_title("Bottom 10 least happy countries")

fig.tight_layout()
plt.show()
```

- The analysis revealed that eight out of the top 10 happiest countries are from the Western Europe region.
- Similarly, the analysis of the top 10 least happy countries showed that eight out of 10 countries are from the Sub-Saharan Africa region.

### in terms of Life Expectancy

```python
fig, ax = plt.subplots(2, 1, figsize = (15,10))

sns.barplot(top_10, y = 'healthy_life_expectancy', x = 'country_name', palette = 'tab20', ax = ax[0])
ax[0].set_title("Top 10 Countries with high Life Expectancy")



sns.barplot(bottom_10, y = 'healthy_life_expectancy', x = 'country_name', palette = 'tab20', ax = ax[1])
ax[1].set_title("Bottom 10 Countries with least Life Expectancy ")

fig.tight_layout()
plt.show()
```

- On an average the top 10 happiest countries Life Expectancy is above 70 years so if you are from one of these countries then you are expected to live for more than 70 years.
- And if you check the bottom 10 least happy countries Life Expectancy you see here most of them are below 60 years of life expectancy age so if you are from these countries then you are expected to live 10 years less then top 10 happiest countries.
